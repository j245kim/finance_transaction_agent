```python
import yfinance as yf
import pandas as pd
import os
from datetime import datetime

# 나스닥 시가총액 상위 50개 기업의 티커 목록 (2025년 2월 기준)
nasdaq_top_50_tickers = [
    'AAPL', 'NVDA', 'MSFT', 'GOOGL', 'AMZN', 'META', 'TSLA', 'AVGO', 'ASML', 'PEP',
    'COST', 'CSCO', 'TMUS', 'TXN', 'ADBE', 'CMCSA', 'AMGN', 'QCOM', 'HON', 'INTC',
    'AMD', 'INTU', 'SBUX', 'AMAT', 'MDLZ', 'BKNG', 'ISRG', 'ADI', 'GILD', 'LRCX',
    'REGN', 'MU', 'VRTX', 'FISV', 'ATVI', 'KLAC', 'MRVL', 'PANW', 'ADP', 'CSX',
    'MELI', 'KDP', 'MAR', 'FTNT', 'MNST', 'CTAS', 'AEP', 'ORLY', 'IDXX', 'LULU'
]

# 현재 날짜 가져오기 (YYYY-MM-DD 형식)
current_date = datetime.today().strftime('%Y-%m-%d')

# 저장할 폴더 경로 설정 (현재 디렉토리의 stock_price_50 폴더)
output_dir = os.path.join(os.getcwd(), "stock_price_50")

# 폴더가 존재하지 않으면 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 주식 데이터 추출 함수
def StockData(ticker, start_date):
    df = ticker.history(start=start_date)
    df = df[['Open', 'High', 'Low', 'Volume', 'Close']]  # 필요한 열만 유지
    return df

# 시작 날짜 설정 (1년 전)
start_date = "2023-01-01"

# 각 티커의 주가 데이터 저장 및 CSV 파일 생성
for ticker_symbol in nasdaq_top_50_tickers:
    try:
        ticker = yf.Ticker(ticker_symbol)
        df = StockData(ticker, start_date)

        # 파일명: 티커이름_YYYY-MM-DD.csv
        file_name = f"{ticker_symbol}_{current_date}.csv"
        file_path = os.path.join(output_dir, file_name)

        # CSV 저장
        df.to_csv(file_path)

        print(f"✅ {ticker_symbol} 데이터 저장 완료: {file_path}")
    
    except Exception as e:
        print(f"❌ {ticker_symbol} 데이터 가져오기 실패: {e}")
```
