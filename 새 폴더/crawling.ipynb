```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import re
import time
import requests

# 크롬 드라이버 경로 설정
chrome_driver_path = r"C:\chromedriver\chromedriver.exe"

# 다운로드 폴더 설정
download_folder = "C:/Reports"
if not os.path.exists(download_folder):
    os.makedirs(download_folder)  # 폴더 없으면 생성

# 크롬 옵션 설정
chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option("detach", True)  # 크롬 창이 자동으로 닫히지 않도록 설정

# 웹드라이버 실행
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# HTML 주소 저장 딕셔너리
html_dict = prev_html_dict if 'prev_html_dict' in globals() else {}

# 기존 데이터가 있는 경우 마지막 CIK 추출
last_cik = None
if html_dict:
    last_key = list(html_dict.keys())[-1]  # 마지막 key 가져오기
    last_cik = last_key[:10]  # 앞 10자리 (CIK 번호)
    print(f"이전 실행 중단 위치 감지: CIK {last_cik}")

# 목표 URL로 이동
url = "https://www.sec.gov/edgar/search/?r=el#/q=10-K&dateRange=10y&category=form-cat1"
driver.get(url)

# 페이지 로드 대기
wait = WebDriverWait(driver, 10)

cnt = 0

# 검색창 찾기
search_box = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/form/div/div[1]/input[2]')))
search_button = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/div/form/div/div[2]/div[5]/div/button[1]')))

# CIK 리스트 순회 (이전 CIK가 있다면 해당 CIK 이후부터 실행)
resume_index = 0
if last_cik:
    resume_index = cik_list.index(int(last_cik)) + 1  # 마지막 실행했던 CIK 다음부터 시작

for cik_raw in cik_list[resume_index:]:
    # 검색창 내용 삭제 후 대기
    search_box.clear()
    
    cnt += 1
    print(f"CIK {cnt} 처리 중...")
    time.sleep(1)  

    # CIK 번호를 10자리로 변환 후 괄호 추가
    formatted_cik = str(cik_raw).zfill(10)
    cik = f"({formatted_cik})"

    print(f"입력할 CIK 값: {cik}")

    # CIK 값 입력
    search_box.send_keys(cik)

    # 검색 버튼 클릭
    search_button.click()

    # 검색 결과 로드 대기
    time.sleep(5)

    collected_links = []  # 수집된 문서 리스트
    max_documents = 10  # 최대 문서 개수
    next_page_xpath = '/html/body/div[3]/div[2]/div[2]/nav/ul/li[12]/a'  # "Next page" 버튼 XPATH

    while len(collected_links) < max_documents:
        try:
            all_links = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a.preview-file')))
            valid_links = [link for link in all_links if "10-K (Annual report)" in link.text and "EX" not in link.text]

            for link in valid_links:
                if len(collected_links) >= max_documents:
                    break

                print(f"10-K 문서 {len(collected_links) + 1} 처리 중...")
                driver.execute_script("arguments[0].scrollIntoView();", link)
                time.sleep(1)
                driver.execute_script("arguments[0].click();", link)
                wait.until(EC.presence_of_element_located((By.ID, "previewer")))
                span_element = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="previewer"]/div/div/div[1]/h4/span[1]')))
                span_text = span_element.text  
                file_title_match = re.findall(r'\((.*?)\)', span_text)
                file_title = " ".join(file_title_match) + ".html" if file_title_match else f"10-K_Report_{len(collected_links) + 1}.html"
                file_title = f"{formatted_cik}_{file_title}"
                print(f"파일 제목: {file_title}")
                ipreviewer_element = wait.until(EC.presence_of_element_located((By.ID, "ipreviewer")))
                ipreviewer_src = ipreviewer_element.get_attribute("src")
                html_dict[file_title] = ipreviewer_src
                collected_links.append(file_title)
                print(f"{file_title} : {ipreviewer_src}")
                close_button = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[4]/div/div/div[1]/button')))
                close_button.click()
                time.sleep(1)

            if len(collected_links) < max_documents:
                try:
                    next_page_button = driver.find_element(By.XPATH, next_page_xpath)
                    driver.execute_script("arguments[0].scrollIntoView();", next_page_button)
                    time.sleep(1)
                    next_page_button.click()
                    time.sleep(5)
                except:
                    print(f"CIK {cik}: 다음 페이지 없음. 종료.")
                    break

        except Exception as e:
            print(f"CIK {cik} 처리 중 오류 발생: {e}")
            break

# 모든 작업 완료 후 브라우저 종료
driver.quit()

# 수집된 HTML 주소 딕셔너리 출력
print("수집된 HTML 주소 딕셔너리:", html_dict)
```

```python
import os
import json
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

# JSON 파일 경로 설정
json_path = r"C:\Users\RMARKET\Desktop\yong\새 폴더\crawing\html_dict.json"

# JSON 파일 열기 및 데이터 로드
with open(json_path, "r", encoding="utf-8") as f:
    html_dict = json.load(f)

# 저장할 기본 폴더 설정
base_download_folder = r"C:\Users\RMARKET\Desktop\yong\새 폴더\crawing"

# Chrome 옵션 설정 (봇 감지 우회)
chrome_options = Options()
chrome_options.add_argument("--headless")  # 창 없이 실행 (봇 감지 회피)
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")  # Selenium 감지 방지
chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")  # 봇 탐지 방지

# Chrome WebDriver 실행
service = Service()
driver = webdriver.Chrome(service=service, options=chrome_options)

# HTML 파일 다운로드 및 저장
for file_name, doc_url in html_dict.items():
    try:
        # 1. 파일 이름에서 CIK 번호(앞 10자리) 추출
        cik_number = file_name.split("_")[0]

        # 2. 해당 CIK 폴더 경로 생성 (없으면 생성)
        save_folder = os.path.join(base_download_folder, cik_number)
        if not os.path.exists(save_folder):
            os.makedirs(save_folder)

        # 3. 저장할 파일 경로 설정
        save_path = os.path.join(save_folder, file_name)

        # 4. SEC 문서 페이지 열기
        print(f"다운로드 중: {doc_url} -> {save_path}")
        driver.get(doc_url)

        # 5. JavaScript 실행 대기 (5초)
        time.sleep(5)

        # 6. 다운로드한 HTML 저장
        with open(save_path, "w", encoding="utf-8") as f:
            f.write(driver.page_source)

        print(f"✅ 저장 완료: {save_path}")

    except Exception as e:
        print(f"❌ 파일 다운로드 오류 ({file_name}): {e}")

# 드라이버 종료
driver.quit()

print("🚀 모든 다운로드 완료!")

```
